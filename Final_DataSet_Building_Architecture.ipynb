```python
#https://www.crummy.com/software/BeautifulSoup/bs4/doc/
#https://selenium-python.readthedocs.io/installation.html#introduction

#1st import: Allows you to launch/initialise a browser.
#2nd import: Allows you to search for things using specific parameters.
#3rd import: Allows you to wait for a page to load.
#4th import: Specify what you are looking for on a specific page in order to determine that the webpage has loaded.
#5th import: Handling a timeout situation.
#6th this module provides O/S dependent functionality


from selenium import webdriver 
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import os
chrome_path = os.path.realpath('chromedriver')
import pandas as pd
import numpy as np
```

```python
# Specifying incognito mode as you launch your browser[OPTIONAL]
option = webdriver.ChromeOptions()
option.add_argument("--incognito")
```

```python
browser = webdriver.Chrome()
#url="https://www.freelancer.com/jobs/building-architecture/1"
url="https://www.freelancer.com/jobs/building-architecture/?results=100"
browser.get(url)
#waitForLoad(driver)
#print(browser.page_source)
```

```python
# Wait 20 seconds for page to load
timeout = 20
try:
    # Wait until the final element [Avatar link] is loaded.
    # Assumption: If Avatar link is loaded, the whole page would be relatively loaded because it is among
    # the last things to be loaded.
    WebDriverWait(browser, timeout).until(EC.visibility_of_element_located((By.XPATH, "//p[@class='JobSearchCard-primary-description']")))
except TimeoutException:
    print("Timed out waiting for page to load")
    browser.quit()
```

```python
# Get all of the titles for the pinned repositories
# We are not just getting pure titles but we are getting a selenium object
# with selenium elements of the titles.
# find_elements_by_xpath - Returns an array of selenium objects.

title=browser.find_elements_by_xpath("//a[@class='JobSearchCard-primary-heading-link']")
#title_link=browser.find_elements_by_xpath("//a[@class='JobSearchCard-primary-heading-link']")
text_element = browser.find_elements_by_xpath("//p[@class='JobSearchCard-primary-description']")
tag_element=browser.find_elements_by_xpath("//a[@class='JobSearchCard-primary-tagsLink']")
#heading=browser.find_elements_by_xpath("//div[@class='JobSearchCard-primary-heading']")
remains_day=browser.find_elements_by_xpath("//span[@class='JobSearchCard-primary-heading-days']")
#Price_usd=browser.find_elements_by_xpath("//div[@class='JobSearchCard-primary-price']")
Price_usd=browser.find_elements_by_xpath("//div[@class='JobSearchCard-secondary-price']")

total_bids=browser.find_elements_by_xpath("//div[@class='JobSearchCard-secondary-entry']")
#avg_bids=browser.find_elements_by_xpath("//div[@class='JobSearchCard-secondary-entry']")
```

```python
# List Comprehension to get the actual repo titles and not the selenium objects.
titles = [x.text for x in title]
#link = [x.text for x in title_link]
text = [x.text for x in text_element]
tags = [x.text for x in tag_element]
#head = [x.text for x in heading]
days = [x.text for x in remains_day]
price = [x.text for x in Price_usd]
totalbids = [x.text for x in total_bids]
```
